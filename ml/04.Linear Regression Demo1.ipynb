## Problem Statement (Linear Regression Demo)
The dataset is a past history of sales based on various advertisement mode. Our task is to build a model that could predict the sames base on TV ad expenses.

```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Step1:  Load data
df = pd.read_csv('/content/drive/MyDrive/edurekaai/_data/samples/Advertising.csv')
df.head()
```

```python
df.info()
```

```python
# Step2: Plot the relationship between TV ads and Sales
## Objective of the plot is to see if there is a linear pattern existing in the data. (i.e. sales increasing with TV ad)
plt.scatter(df['TV'], df['Sales'])
plt.xlabel('TV Ads')
plt.ylabel('Sales')
plt.show()
```

```python
# Step3: Segregate Independent and Dependent Variables
## and Remove the header from the data
## .values converts the Series to a NumPy array (type: numpy.ndarray) — which is required for most ML libraries like scikit-learn.
X = df["TV"].values
y = df["Sales"].values
print(type(X))
print(type(y))

# This reshapes the arrays into 2D, which is the expected input shape for scikit-learn models (like LinearRegression().fit(X, y)):
## df["TV"] is Panda Series
## df["TV"].values is a NumPy Array
## df["TV"].values.reshape(-1,1) is 2D array.
x = df["TV"].values.reshape(-1,1)
y = df["Sales"].values.reshape(-1,1)
print(x)
# print(y)
```

```python
# Train the Algorithm to create the model
model = LinearRegression()
model.fit(x, y)
```

```python
# Make a prediction from the x-values
y_pred = model.predict(x)
print(y_pred)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('TV Ads')
plt.ylabel('Sales')
plt.show()
```

```python
## Co-efficients
print(model.coef_) #slop value
print(model.intercept_) #y-intercept
```

```python
## Evaluate the model
#print('Mean Absolute Error - MAE:', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error - MSE:', metrics.mean_squared_error(y, y_pred))
print('Root Mean Squared Error - RMSE:', np.sqrt(metrics.mean_squared_error(y, y_pred)))
print('R2 Score:', metrics.r2_score(y, y_pred))

```

```python
## Predict on new data
tv_add = int(input('Enter the TV ads amount: '))
print('Predicted Sales: ', model.predict([[tv_add]]))
```
