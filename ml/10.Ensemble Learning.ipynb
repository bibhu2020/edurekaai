**Ensemble learning** is a machine learning technique that combines the predictions of multiple models (called **base learners** or **weak learners**) to produce a more accurate and robust result than a single model could.

### 1. ðŸ“¦ Bagging (Bootstrap Aggregating)

Bagging creates multiple versions of a model **in parallel** by training on **random subsets** of the dataset (with replacement). The results of these models are combined by **voting** (classification) or **averaging** (regression).

- **Goal:** Reduce variance
- **Common algorithm:** Random Forest

### 2. ðŸš€ Boosting
Boosting trains models **sequentially**, where each model focuses on fixing the errors made by the previous ones. It gives higher importance (weight) to incorrectly predicted instances.

- **Goal:** Reduce bias

- **Popular algorithms**: AdaBoost, Gradient Boosting, XGBoost, LightGBM
