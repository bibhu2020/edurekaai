### Core Components of ML

**1. Representation**: The format in which the data is processed. (e.g. vectors, matrics, graph).

**2. Evaluation**: Metrics used to assess model performance (e.g. accuracy, RMSE, precision, recall).

**3. Optimization**: Techniques to improve the models (e.g gradient descent, hyperparameter tunning).*italicized text*

### Machine Learning Key Terms

**Features**: Independent vaiables or attributes used for prediction. (e.g. input data columns)

**Labels**: Target values in supervised learning.

**Model**: A mathematical representation trained on data data to make predictions.

**Training Data**: The dataset used to train the ML model
