## Data Types

#### Qualitative Data (Catagorical Data)

**Nominal**: Data that represents categories with no inherent order or ranking. 
Examples:
Gender: Male, Female, Non-binary
Country: USA, India, Japan

There is ranking in this data. E.g. Male > Female is not valid statement.

**Ordinal**: Data that represents categories with a meaningful order, but the differences between them are not measured or consistent.
Examples:
Education level: High School < Bachelor's < Master's < PhD
Satisfaction: Unsatisfied < Neutral < Satisfied

#### Quntitative Data (Numerical Data)
Anything that can be measured, aggregrated, and summerized.

**Discrete**: Anything that is countable. E.g. no of students in the class.

**Continious**: Something that can't be counted. E.g. amount of rain. It is not whole number. 





## Variables
A variable is an algebric representation of a unknown value or a value that varies.

### Type of Variables

**Categorical Variable** : A variable that represents categorical data. 

**Control Variable**: A factor that must be held constant in an experimental research. 

**Confounding Variable**: Extra hidden variables that have an impact or bias on the observation. **_As a researcher, you must eliminate the confounding variables._**

**Depdendent Variable**:
E.g. Salary of an Employee is determined by his no of years of exprience, and qualification. In this case, Salary is dependent variable

**Indepdendent Variable**: Exprience and Qualifications are Independent variables.

***Results from non-experimental research (observation data), inferences are co-relation.

***Results from experimental research, inferences are causations. 

# Sampling Techniques

Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population. It is especially useful in statistics and research where studying an entire population is impractical due to time, cost, or logistics.

Sampling techniques can be broadly classified into **Probability Sampling** and **Non-Probability Sampling**.

---

## Probability Sampling (or Random Sampling)

In **probability sampling**, every member of the population has a known and equal chance of being selected. This technique ensures unbiased representation of the population.

### **Random Sampling**
Each individual is chosen entirely by chance, and each member of the population has an equal chance of being included.

**Example:**  
You want to survey students in a university with 10,000 students. You randomly select 500 student IDs using a random number generator.

**Pros:**
- Eliminates selection bias.
- Easy to analyze statistically.

**Cons:**
- Requires a complete list of the population.
- Not feasible for very large populations without digital tools.

---

### **Systematic Sampling**
Involves selecting every *k-th* individual from a list of the population after randomly selecting a starting point.

**Example:**  
From a list of 1,000 employees, you pick every 10th name starting from the 5th person.

**Pros:**
- Simple and quick to implement.
- Easier than random sampling.

**Cons:**
- Can introduce bias if the list has a hidden pattern.
- Less random than simple random sampling.

---

### **Stratified Sampling**
The population is divided into strata (groups) based on a specific characteristic, and random samples are taken from each group proportionally.

**Example:**  
A school has 60% girls and 40% boys. To survey 100 students, you randomly pick 60 girls and 40 boys.

**Pros:**
- Ensures representation across key subgroups.
- Increases statistical precision.

**Cons:**
- Requires detailed knowledge of population structure.
- More complex to organize.

---

## Non-Probability Sampling (Non-Random Sampling)

In **non-probability sampling**, not every member has a chance of being selected. These methods are more prone to bias and often used in exploratory research.

### **Snowball Sampling**
Existing study subjects recruit future subjects from among their acquaintances.

**Example:**  
For a study on drug addiction, you start with a few known users who refer you to other users in their network.

**Pros:**
- Useful for hard-to-reach or hidden populations.
- Cost-effective.

**Cons:**
- High risk of selection bias.
- Limits generalizability.

---

### **Quota Sampling**
The population is segmented, and the researcher fills quotas for each segment (not randomly).

**Example:**  
You need 50 male and 50 female participants. You survey people until the quota for each gender is filled.

**Pros:**
- Ensures representation of key subgroups.
- Faster and cheaper than stratified sampling.

**Cons:**
- Subject to selection bias.
- Not statistically generalizable.

---

### **Judgment (or Purposive) Sampling**
Subjects are selected based on the researcher's judgment about who will be the most useful or representative.

**Example:**  
A researcher selects five experienced managers to understand corporate decision-making.

**Pros:**
- Useful for expert opinions.
- Focused and relevant data.

**Cons:**
- Subjective and potentially biased.
- Cannot infer population-level conclusions.

---

### **Convenience Sampling**
Samples are taken from a group that is conveniently accessible to the researcher.

**Example:**  
You survey people walking by on the street outside your office.

**Pros:**
- Very easy and low-cost.
- Good for pilot studies.

**Cons:**
- Highly biased.
- Not representative of the population.

---

## Conclusion

Choosing the right sampling technique depends on the research goals, available resources, and the nature of the population. While probability sampling is ideal for generalizing findings, non-probability sampling is often practical for exploratory or qualitative research.
