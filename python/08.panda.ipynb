```python
# Fix missing data

## According to statistics
## Whenever you have any data which categorical in nature and missing, then identify the MODE of it and replace with them.
### MODE means most repeated value
## Whenever you have any data which is numerical in nature and missing, then identity the mean/median and replace with them.
### Caution: Mean is affected by outliers.  Median is not sensitive to outliers. Hence, use median incase you have outliers.
```

```python
print(f"Mode: {df_bigmart.Item_MRP.mode()[0]}") #there could be multiple outliers
print(f"Mode: {df_bigmart.Item_MRP.mode()[1]}") #there could be multiple outliers
print(f"Mean: {df_bigmart.Item_MRP.mean()}")
print(f"Median: {df_bigmart.Item_MRP.median()}")
```

```python
## Handling multiple data frame (data may be split into 2 data frames)
df_bigmart.columns

# Creating 2 data frames to simulate the scenario of data coming from 2 sources.
df_bigmart1 = df_bigmart[['Item_ID', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility','Item_Type', 'Item_MRP', 'Outlet_ID', 'Outlet_Establishment_Year']]
df_bigmart2 = df_bigmart[['Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type','Item_Outlet_Sales']]

# Concartenate by columns.
pd.concat([df_bigmart1, df_bigmart2], axis=1)
```

```python
# Joining 2 dataframes using joining KEY
# Created 2 more data frame with a common key
df_bigmart1_1 = df_bigmart1[['Item_ID', 'Item_Weight', 'Item_Fat_Content']].head()
df_bigmart1_2 = df_bigmart1[['Item_ID', 'Item_MRP']].head()

df_bigmart1_2
```

```python
##change 2 row KEY value
df_bigmart1_1.loc[3, 'Item_ID'] = 'G1003'
df_bigmart1_1.loc[1, 'Item_ID'] = 'G1001'

## Inner Join
pd.merge(df_bigmart1_1, df_bigmart1_2, on=["Item_ID"], how="inner")

## left Join (inner + left)
pd.merge(df_bigmart1_1, df_bigmart1_2, on=["Item_ID"], how="left")

## right Join (inner + right)
pd.merge(df_bigmart1_1, df_bigmart1_2, on=["Item_ID"], how="right")

## outer Join (inner + left over from left + left over from right)
pd.merge(df_bigmart1_1, df_bigmart1_2, on=["Item_ID"], how="outer")
```

### Testing Dataframe Knowledge (using stock data)

```python
df_stocks = pd.read_csv('../_data/samples/stock_trending_data.csv')
df_stocks.head()
```

```python
# Find highest closing price of each stock
df_stocks.groupby('Symbol')['Close'].max()
```

```python
# Calculate daily change in stock price (close - open)
df_stocks['Daily_Change'] = (df_stocks['Close'] - df_stocks['Open']) * 100 / df_stocks['Open']

df_stocks.head()
```

```python
# Filter high volume trading days
df_stocks['Mean_Volume'] = df_stocks['Volume'].mean()
df_stocks[df_stocks['Volume'] > df_stocks['Mean_Volume']].head()
```
