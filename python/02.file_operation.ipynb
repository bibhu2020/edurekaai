```python
# File operations in python
# File - Basic persistence storage (text files)
# File does not include MS Word, Excel, PDF etc. (specialized formats)
# Operations:  read, write, append, delete, open, close
```

```python
# fp = open("../_data/samples/test.txt", "w")  # open file in write mode
fp = open("/content/drive/MyDrive/edurekaai/_data/samples/test.txt", "w")  # open file in write mode
print(fp)  # print file object
print(fp.name)  # print file name
print(fp.mode)  # print file mode
print(fp.closed)  # print file closed status
print(fp.encoding)  # check if file is readable
```
