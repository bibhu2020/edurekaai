```python
# Regular expression module
import re

string1 = "Python is a suprt language but python has its own advantages. Python is being used in machine learning."

# It is case sensitive. Wherever it find the first match, it will stop searching. It returns a match object.
print(re.search("Python", string1))

# It finds all matches. It will return a list of match strings.
print(re.findall("Python", string1))  # It will find all the matches

# it matches the string "python" or "Python" in a case insensitive manner.
print(re.findall("[pP]ython", string1))  # It will find all the matches

# It will replace the first match with Java= sys.path
print(re.sub('[pP]ython', 'Java', string1))  # It will replace all the matches with Java

string2 = "Customer request received @12:40 - 123.333.123.123 - Ram Kumar - 1234567890"

# This is an advanced split where, it will split the string at @, :, and -
print(re.split('[@:-]', string2))  # It will split the string at @
```

```python
# Using regular expression to read log file
import re
import os
print(os.getcwd())
print(os.path.exists("../_data/samples/samplelog.txt"))

reg_object = re.compile('[@:-]')

fp = open('../_data/samples/samplelog.txt')
for line in fp:
    line = line.strip()
    #result = re.split('[@:-]', line) # this compiles the regex everytime
    result = reg_object.split(line)  # this compiles the regex only once
    # print(result[1:])
    print(list(map(str.strip, result[1:])))
```
