# Apriori Algorithm

```python
from apyori import apriori
```

## Preprocessing

```python
# for apriori, you will have to read file row by row
fs = open('/content/drive/MyDrive/edurekaai/_data/samples/store_data.csv', 'rt')
raw_data = fs.readlines()
fs.close()
raw_data[:5]
```

```python
# Convert the data to a 2D array (list of a list)
store_data = [line.strip().split(',') for line in raw_data]
print(store_data[:5]) # [[a,b,c],[d,a,c],[c,d], [b,a,e,d]]
print(len(store_data)) # 7501

# Print first 10 rows (or fewer if not enough rows)
print("First 5 rows:")
for i, row in enumerate(store_data[:5]):
    print(f"Row {i+1}: {row}")
```

```python
# Apply Apriori
# Keep changing the min_support and min_confidence to find solid association rules
rules = apriori(store_data, min_support=0.006, min_confidence=0.30, min_lift=3, min_length=3)

rules_list = list(rules)
print("Total Rules: ", len(rules_list))
print(type(rules_list))
for item in rules_list[:10]:
  pair = item[0]
  items = [x for x in pair]
  print("Association Rule: " + items[0] + " -> " + items[1])
  print("Support: " + str(item[1]))
  print("Confidence: " + str(item[2][0][2]))
  print("Lift: " + str(item[2][0][3]))
  print("=====================================")
```
